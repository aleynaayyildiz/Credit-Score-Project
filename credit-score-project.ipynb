<p>Kredi Skoru Sınıflandırma projesi, kredi skorlarını sınıflandırmak için sağlam ve doğru bir sistem geliştirmeyi amaçlamaktadır. Bu sistem, bireylerin kredi değerliliğini daha etkin bir şekilde değerlendirmede ve kredi kartları, krediler ve diğer finansal hizmetler için uygunluklarını belirlemede yardımcı olacaktır. Kredi skoru sınıflandırması, finansal kurumlar için risk değerlendirmesinin kritik bir bileşeni olarak, temerrüt riskini en aza indirirken bilinçli kararlar almalarına olanak tanıyacaktır. Bu proje, finansal kuruluşların daha güvenilir ve doğru kredi değerlendirmeleri yapmasına yardımcı olacak, böylece hem müşteriler hem de kurumlar için daha güvenli ve sağlam bir finansal ortam oluşturulacaktır. </p>

Problemi Anlama
Küresel bir finans şirketi olan XYZ Finans, dünya genelinde milyonlarca müşteriye hizmet vermektedir. Şirket, kredi kartları, kişisel krediler ve çeşitli finansal ürünler sunmaktadır. Bu hizmetleri sunarken, müşterilerin kredi değerliliğini doğru bir şekilde belirlemek hayati önem taşımaktadır. Ancak, mevcut manuel kredi değerlendirme süreçleri hem zaman alıcı hem de hata yapma olasılığı yüksek olduğundan, daha verimli ve güvenilir bir çözüm arayışına girilmiştir.

Güçlü veri altyapısı ve ileri teknoloji kullanımıyla dikkat çeken XYZ Finans, müşteri memnuniyetini en üst düzeyde tutmayı hedeflemektedir.

Sorun Tanımı
Şirket, yıllar boyunca topladığı geniş veri setleri sayesinde, müşterilerinin krediyle ilgili bilgilerine sahiptir. Ancak, bu verilerin manuel olarak işlenmesi ve kredi puanı değerlendirmelerinin yapılması hem büyük bir iş yükü yaratmakta hem de potansiyel hata payını artırmaktadır. Yönetim, bu süreci daha etkili ve hızlı hale getirmek amacıyla, müşterilerin kredi bilgilerini kullanarak kredi puanlarını otomatik olarak sınıflandırabilecek bir sistem geliştirmek istemektedir.

Proje Hedefi
Bu projenin temel amacı, müşterilerin krediyle ilgili bilgilerini kullanarak kredi puanlarını doğru bir şekilde sınıflandırabilecek bir makine öğrenimi modeli geliştirmektir. Bu sistem, kredi değerliliği değerlendirme sürecini otomatikleştirerek, manuel çabaları azaltacak ve değerlendirmelerin doğruluğunu artıracaktır.  <span style="color:#2c3e50;font-weight: bold;font-size: 18px "><span style="font-size: 25px;"> 📌 </span>  "column_details" fonksiyonu ile df datafremimizdeki belirtilen bir sütunun; <br><br> ⚫ Veri tipini, <br><br> ⚫ Null değeri olup olmadığını,<br><br>  ⚫ Benzersiz değerleri ve bu değerlerin sayısını, <br><br> ⚫ Yüzdelik dağılımlarını hesaplayan bir fonksiyon oluşturuyoruz. <br><br> Bu fonksiyon, df datafremimizdeki herhangi bir sütun için tekrar tekrar aynı işlemleri yapmanıza gerek kalmadan, sütunun istediğimiz  bilgilerini hızlıca elde etmemizi sağlar. <br><br></span>

```python
def column_details(df, column):
    """
    Belirtilen sütunun detaylarını hesaplayan fonksiyon.
    
    Parametreler:
    df : Analiz edilecek DataFrame.
    column_name : Analiz edilecek sütunun adı.
    
    """
    
    print( column, "Kolonunun Detayları")
    
    # Sütunun veri tipini yazdırır
    # df[column].dtype ifadesi, DataFrame'deki (df) belirtilen sütunun veri tipini döndürür.
    print("\nVeri Tipi: ", df[column].dtype)
    
    # Null (boş) değerlerin olup olmadığını kontrol eder
    # df[column].isnull().sum() ifadesi, belirtilen sütunda kaç adet null değer olduğunu sayar.
    
    count_null = df[column].isnull().sum()
    if count_null == 0:
        # Eğer null değer yoksa, bunu yazar
        print("\nBoş değer bulunmamaktadır.")
    elif count_null > 0:
        # Eğer null değer varsa, kaç adet olduğunu yazar
        print("\n ", count_null, " tane boş değer bulunmaktadır.")  
        
        
    # Benzersiz değerlerin sayısını alır
    # df[column].nunique() ifadesi, belirtilen sütundaki benzersiz değerlerin sayısını döndürür.
    print("\nBenzersiz Değerlerin Sayısı: ", df[column].nunique())
    
    # Benzersiz değerleri getirir
    print("\nBenzersiz Değerler: ", df[column].unique())
    
   # column değişkeninin yüzdelerini hesaplama
   # df[column]: df DataFrame'inden column sütununu seçer.
   # .value_counts(normalize=True): Bu sütundaki benzersiz değerlerin frekanslarını hesaplar 
   # ve bunları toplam içindeki oranlar olarak döndürür.
   # normalize=True: Her bir benzersiz değer sayısının toplam veri içindeki yüzdesini hesaplar. 
   # Bu, her değerin toplam gözlemler içindeki oranını verir.
   # * 100: Her bir oranı 100 ile çarparak yüzdelik değerlere dönüştürür.
    print("\nKolonun Dağılımı:\n")
    print( df[column].value_counts(normalize=True) * 100)
```

```python
def statistics_numerical_column(df, column_name):
    """
    Verilen sayısal sütunun özet istatistiklerini hesaplar ve sonucu normal biçimde gösterir.
    
    NOT: Eğer CV değeri yüzde 35'in altında ise, veri setinin normal dağılıma daha yakın olabileceği
    ve aykırı değerlerin az olduğu söylenebilir. 
    Eğer CV değeri yüzde 35'in üzerinde ise, veri setinde büyük bir değişkenlik olduğu 
    ve aykırı değerlerin bulunma olasılığının yüksek olduğu anlamına gelir.
    
    Args:
    df : Veri seti
    column_name : İşlenecek sütunun adı
    
    """
   
    # Verilen sütunun özet istatistiklerini gösterme
    column_description = df[[column_name]].describe()

    print(f"'{column_name}' sütununun özet istatistikleri:")
    print(column_description)

    # mean ve std değerlerini çekmek için doğru indeksleme
    count = column_description.loc['count'][column_name]
    mean = column_description.loc['mean'][column_name]
    std = column_description.loc['std'][column_name]
    max_ = column_description.loc['max'][column_name]
    min_ = column_description.loc['min'][column_name]
    percentile_25 = column_description.loc['25%'][column_name]
    percentile_75 = column_description.loc['75%'][column_name]
    percentile_50 = column_description.loc['50%'][column_name]

    # Değişim katsayısını (CV) hesaplama
    cv = (std / mean) * 100  # Yüzde olarak ifade edilir
    
    print(f"\nCount: {count:,.0f}")
    print(f"Mean: {mean:,.0f}")
    print(f"Standard Deviation (Std): {std:,.0f}")
    print(f"Minimum (Min): {min_:,.0f}")
    print(f"25th Percentile: {percentile_25:,.0f}")
    print(f"50th Percentile (Median): {percentile_50:,.0f}")
    print(f"75th Percentile: {percentile_75:,.0f}")
    print(f"Maximum (Max): {max_:,.0f}")
    print(f"\nDeğişim Katsayısı (CV): {cv:.2f}%")


    # CV değerini kontrol etme
    if cv < 35:
        print("Değişim katsayısı %35'in altında. Veri normal dağılımda olabilir ve aykırı değerler azdır.")
    else:
        print("Değişim katsayısı %35'in üzerinde. Veri aykırı değerler bulunmaktadır.")
```

```python
def detect_outliers_iqr(data, column_name, Q1=0.25, Q3=0.75):
    # Aykırı değerleri tespit etmek için IQR yöntemini kullanma
    Q1 = df[column_name].quantile(Q1)
    Q3 = df[column_name].quantile(Q3)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)]
    
    # Aykırı değerleri küçükten büyüğe sıralama
    sorted_outliers = outliers.sort_values(by = column_name, ascending=False)
    
    print("Aykırı Değerler (Küçükten Büyüğe):")
    print(sorted_outliers[[column_name]].head(16))
    
    # Aykırı değerlerin sayısını bulma
    num_outliers = len(outliers)
    
    print("Aykırı Değerlerin Sayısı:")
    print(num_outliers)
    
    # Aykırı değerlerin benzersiz (unique) değerlerini alma
    unique_outliers = sorted_outliers[column_name].unique()
    
    print("Benzersiz Aykırı Değerler:")
    print(unique_outliers)
```

```python
def plot_countplot(df,column,rotation=0):
    palette = "deep" 
    sns.set_palette(palette)
    
    sns.countplot(data=df, x=column)

    plt.xlabel(f'{column}')
    plt.ylabel('Kayıt Sayısı')
    plt.title(f'{column} Dağılımı')
    plt.xticks(rotation=rotation)

    plt.show()
```

```python
def plot_stacked_bar(df,column1,column2,rotation=0):
    print(f'\n{column1} & {column2} Dağılımı')
    palette = "deep" 
    sns.set_palette(palette)

    pd.crosstab(df[column1], df[column2]).plot(kind='bar', stacked=True)
    
    plt.xlabel(f'{column1}')
    plt.ylabel('Kayıt Sayısı')
    plt.title(f'{column1} & {column2} Dağılımı')
    plt.xticks(rotation=rotation)

    plt.show()
```

```python
def plot_displot(df,column,rotation=0,bins=20):
    print("\n-----------------------------------------------------")
    print(f'\n{column} Dağılımı')
    palette = "deep" 
    sns.set_palette(palette)
    
    sns.displot(data=df, x=column, kde=True, bins=bins)

    plt.xlabel(f'{column}')
    plt.ylabel('Kayıt Sayısı')
    plt.title(f'{column} Dağılımı')
    plt.xticks(rotation=rotation)

    plt.show()
```

```python
def numeric_cleaning(df, col, groupby):    
    df_dropped = df[df[col].notna()].groupby(groupby)[col].apply(list)

    # Gruplandırılmış 'N_Age' listeleri üzerinde mode (mod) hesapla 
    # x'e min'i y'ye max'ı ata 
    x, y = df_dropped.apply(lambda x: stats.mode(x, keepdims=True)).apply([min, max])

    # Her bir müşterinin 'N_Age' listesi için minimum ve maksimum mod değerlerini al
    minn, maxx = x[0][0], y[0][0]
    
    # Yanlış değerlere Nan ata
    df[col] = df[col].apply(lambda x: np.nan if ((x<minn)|(x>maxx)|(x<0)) else x)
    
    # Local Mode ile Nan doldur
    mode_by_group = df.groupby(groupby)[col].transform(lambda x: x.mode()[0] if not x.mode().empty else np.NaN)
    df[col] = df[col].fillna(mode_by_group)
    
    print("\nTemizlendikten sonra Min, Max değerleri:", df[col].apply([min, max]), sep='\n', end='\n') 
    print("\nTemizlendikten sonra eşsiz değerlerin sayısı:",df[col].nunique())
    print("\nTemizlendikten sonra boş değer sayısı:",df[col].isnull().sum())
```

     ⚫ "N_Type_of_Loan" sütunundaki tüm değerler incelendi ve bu değerlerin içindeki benzersiz alt         değerler (loan <br>türlerini) bulundu. Bu benzersiz alt değerler için ayrı ayrı sütunlar oluşturuldu ve bu sütunlarda 1 veya 0 değerlerini<br>  kullanarak her satırın ilgili değeri içerip içermediğini belirlendi. Bu işlem, her benzersiz loan türü için bir sütun ekleyerek, satırların bu loan türlerini içerip içermediğini ikili (binary) değerlerle gösterir. Bu sayede, kategorik <br>verilerimizi daha analiz edilebilir hale getirmiş olduk.<br>

```python
# Benzersiz loan türlerini saklamak için boş bir set oluşturuldu
unique_values = set()

""" 
    Ek Bilgi: 
    ----------------------------------------
    "append" metodu:
    Kullanım Alanı: Listeler için kullanılır.
    Fonksiyon: Tek bir öğeyi listeye ekler.
    
    ---------------------------------------
    "extend" metodu:
    Kullanım Alanı: Listeler için kullanılır.
    Fonksiyon: Bir listeye başka bir listenin öğelerini teker teker ekler.
    
    ---------------------------------------
    "update" metodu:
    Kullanım Alanı: Setler için kullanılır.
    Fonksiyon: Bir iterable'dan (örneğin liste, tuple) öğeleri sete teker teker ekler.
    
    ----------------------------------------
    "lambda" ifadesi içinde " unique_values.update(x.split(",")) " ifadesi çalıştırılıyor.
     Bu ifade None döndürüyor çünkü set.update metodu bir değeri döndürmez; sadece seti günceller.
     Bu nedenle, lambda ifadesinin döndürdüğü sonuç None olur.

"""
# dropna() metodu ile N_Type_of_Loan sütunundaki NaN değerler atlanır.
# apply metodu ile her bir hücredeki değer, lambda fonksiyonuna geçirilir.
# Lambda fonksiyonunda x.split(",") ifadesi ile hücredeki değer virgülle ayrılır ve bir liste döner.
# strip() metodu ile listenin her bir öğesindeki sağ ve sol boşluklar temizlenir.
# unique_values.update() metodu ile bu temizlenmiş öğeler unique_values setine eklenir.

df["N_Type_of_Loan"].dropna().apply(lambda x: unique_values.update([item.strip() for item in x.split(",")]))
```

```python
# Boş değeri (empty string) 'unique_values' setinden çıkarıldı
unique_values.discard('')

# 'unique_values' setini bir listeye dönüştürüldü
unique_values_list = list(unique_values)
```

```python
""" 
    for value in unique_values_list:: unique_values_list içerisindeki her değer için döngü başlatılır.
    
    df[value] = df["N_Type_of_Loan"].apply(...): N_Type_of_Loan sütununda her bir hücreyi kontrol 
    ederek, value değerini içeren hücreler için yeni bir sütun oluşturur ve
    bu sütuna 1, içermeyenlere 0 atar.

    lambda x: 1 if pd.notna(x) and value in [item.strip() for item in x.split(",")] else 0: 
    Lambda fonksiyonu, hücredeki değerin NaN olup olmadığını ve value içeriğini kontrol eder. 
    Eğer hücre NaN değilse ve value içeriği bulunuyorsa 1, aksi takdirde 0 döner.
    Hücredeki değeri virgülle ayırarak kontrol eder ve her bir öğenin başındaki ve 
    sonundaki boşlukları temizler.

"""
for value in unique_values_list:
    df[value] = df["N_Type_of_Loan"].apply(lambda x: 1 if pd.notna(x) and value in [item.strip() for item in x.split(",")] else (0 if pd.notna(x) else np.nan))

```

# Ki Kare Bağımsızlık Testi
İki kategorik değişken arasındaki bağımsızlığı test etmek için kullanılır.

```python
import scipy.stats as ss  # scipy.stats kütüphanesini ss olarak içe aktar

# 1. Adım: Kategorik değişkenlerin seçimi
cat_column = ["Month", "Category_Age", "N_Occupation", "N_Type_of_Loan", "Payment_of_Min_Amount",
              "N_Payment_Behaviour", "N_Credit_Mix", "Credit_Status"]

# 2. Adım: P-Değerlerini İçeren Boş Bir Matris Oluşturma
p_matrix = pd.DataFrame(index=cat_column, columns=cat_column)


for col1 in cat_column:
    for col2 in cat_column:
        if col1 == col2:
            p_matrix.loc[col1, col2] = np.nan  # Kendisiyle karşılaştırma yok
        else:
            contingency_table = pd.crosstab(dnew[col1], dnew[col2])
            # Hücre beklenen frekanslarının kontrolü
            chi2, p, dof, ex = ss.chi2_contingency(contingency_table)
            if np.all(ex >= 5):
                p_matrix.loc[col1, col2] = p
            else:
                p_matrix.loc[col1, col2] = np.nan  # Beklenen frekanslar yeterli değilse NaN ata
  

# P-Değerlerini Sayısal Değere Dönüştürme
p_matrix = p_matrix.astype(float)


# P-Değerlerini Görselleştirme
plt.figure(figsize=(10, 8))
sns.heatmap(p_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Categorical Features Chi-Square Test p-Values')
plt.show()
```

# ikili kategorik bağımlı değişken ile sayısal bağımsız değişkenler arasındaki ilişkiyi Mann-Whitney U testi ile nasıl analiz edeceğinizi 

```python
from scipy.stats import mannwhitneyu
# Sayısal değişkenleri seçme
numerical_columns = dnew.select_dtypes(include=[np.number]).columns

# P-değerlerini saklamak için boş bir DataFrame oluşturma
p_values = pd.DataFrame(index=numerical_columns, columns=['p_value'])

# Her sayısal değişken için Mann-Whitney U Testi uygulama
for col in numerical_columns:
    group1 = dnew[dnew['Credit_Status'] == 'yes'][col]
    group2 = dnew[dnew['Credit_Status'] == 'no'][col]
    stat, p = mannwhitneyu(group1, group2)
    p_values.loc[col, 'p_value'] = p

# P-değerlerini sayısal hale getirme
p_values = p_values.astype(float)

# P-değerlerini ısı haritası olarak görselleştirme
plt.figure(figsize=(8, 6))
sns.heatmap(p_values, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Mann-Whitney U Testi P-Değerleri Isı Haritası')
plt.show()

print(p_values)
```
