```python
def column_details(df, column):
    """
    Belirtilen sÃ¼tunun detaylarÄ±nÄ± hesaplayan fonksiyon.
    
    Parametreler:
    df : Analiz edilecek DataFrame.
    column_name : Analiz edilecek sÃ¼tunun adÄ±.
    
    """
    
    print( column, "Kolonunun DetaylarÄ±")
    
    # SÃ¼tunun veri tipini yazdÄ±rÄ±r
    # df[column].dtype ifadesi, DataFrame'deki (df) belirtilen sÃ¼tunun veri tipini dÃ¶ndÃ¼rÃ¼r.
    print("\nVeri Tipi: ", df[column].dtype)
    
    # Null (boÅŸ) deÄŸerlerin olup olmadÄ±ÄŸÄ±nÄ± kontrol eder
    # df[column].isnull().sum() ifadesi, belirtilen sÃ¼tunda kaÃ§ adet null deÄŸer olduÄŸunu sayar.
    
    count_null = df[column].isnull().sum()
    if count_null == 0:
        # EÄŸer null deÄŸer yoksa, bunu yazar
        print("\nBoÅŸ deÄŸer bulunmamaktadÄ±r.")
    elif count_null > 0:
        # EÄŸer null deÄŸer varsa, kaÃ§ adet olduÄŸunu yazar
        print("\n ", count_null, " tane boÅŸ deÄŸer bulunmaktadÄ±r.")  
        
        
    # Benzersiz deÄŸerlerin sayÄ±sÄ±nÄ± alÄ±r
    # df[column].nunique() ifadesi, belirtilen sÃ¼tundaki benzersiz deÄŸerlerin sayÄ±sÄ±nÄ± dÃ¶ndÃ¼rÃ¼r.
    print("\nBenzersiz DeÄŸerlerin SayÄ±sÄ±: ", df[column].nunique())
    
    # Benzersiz deÄŸerleri getirir
    print("\nBenzersiz DeÄŸerler: ", df[column].unique())
    
   # column deÄŸiÅŸkeninin yÃ¼zdelerini hesaplama
   # df[column]: df DataFrame'inden column sÃ¼tununu seÃ§er.
   # .value_counts(normalize=True): Bu sÃ¼tundaki benzersiz deÄŸerlerin frekanslarÄ±nÄ± hesaplar 
   # ve bunlarÄ± toplam iÃ§indeki oranlar olarak dÃ¶ndÃ¼rÃ¼r.
   # normalize=True: Her bir benzersiz deÄŸer sayÄ±sÄ±nÄ±n toplam veri iÃ§indeki yÃ¼zdesini hesaplar. 
   # Bu, her deÄŸerin toplam gÃ¶zlemler iÃ§indeki oranÄ±nÄ± verir.
   # * 100: Her bir oranÄ± 100 ile Ã§arparak yÃ¼zdelik deÄŸerlere dÃ¶nÃ¼ÅŸtÃ¼rÃ¼r.
    print("\nKolonun DaÄŸÄ±lÄ±mÄ±:\n")
    print( df[column].value_counts(normalize=True) * 100)
```

```python
def statistics_numerical_column(df, column_name):
    """
    Verilen sayÄ±sal sÃ¼tunun Ã¶zet istatistiklerini hesaplar ve sonucu normal biÃ§imde gÃ¶sterir.
    
    NOT: EÄŸer CV deÄŸeri yÃ¼zde 35'in altÄ±nda ise, veri setinin normal daÄŸÄ±lÄ±ma daha yakÄ±n olabileceÄŸi
    ve aykÄ±rÄ± deÄŸerlerin az olduÄŸu sÃ¶ylenebilir. 
    EÄŸer CV deÄŸeri yÃ¼zde 35'in Ã¼zerinde ise, veri setinde bÃ¼yÃ¼k bir deÄŸiÅŸkenlik olduÄŸu 
    ve aykÄ±rÄ± deÄŸerlerin bulunma olasÄ±lÄ±ÄŸÄ±nÄ±n yÃ¼ksek olduÄŸu anlamÄ±na gelir.
    
    Args:
    df : Veri seti
    column_name : Ä°ÅŸlenecek sÃ¼tunun adÄ±
    
    """
   
    # Verilen sÃ¼tunun Ã¶zet istatistiklerini gÃ¶sterme
    column_description = df[[column_name]].describe()

    print(f"'{column_name}' sÃ¼tununun Ã¶zet istatistikleri:")
    print(column_description)

    # mean ve std deÄŸerlerini Ã§ekmek iÃ§in doÄŸru indeksleme
    count = column_description.loc['count'][column_name]
    mean = column_description.loc['mean'][column_name]
    std = column_description.loc['std'][column_name]
    max_ = column_description.loc['max'][column_name]
    min_ = column_description.loc['min'][column_name]
    percentile_25 = column_description.loc['25%'][column_name]
    percentile_75 = column_description.loc['75%'][column_name]
    percentile_50 = column_description.loc['50%'][column_name]

    # DeÄŸiÅŸim katsayÄ±sÄ±nÄ± (CV) hesaplama
    cv = (std / mean) * 100  # YÃ¼zde olarak ifade edilir
    
    print(f"\nCount: {count:,.0f}")
    print(f"Mean: {mean:,.0f}")
    print(f"Standard Deviation (Std): {std:,.0f}")
    print(f"Minimum (Min): {min_:,.0f}")
    print(f"25th Percentile: {percentile_25:,.0f}")
    print(f"50th Percentile (Median): {percentile_50:,.0f}")
    print(f"75th Percentile: {percentile_75:,.0f}")
    print(f"Maximum (Max): {max_:,.0f}")
    print(f"\nDeÄŸiÅŸim KatsayÄ±sÄ± (CV): {cv:.2f}%")


    # CV deÄŸerini kontrol etme
    if cv < 35:
        print("DeÄŸiÅŸim katsayÄ±sÄ± %35'in altÄ±nda. Veri normal daÄŸÄ±lÄ±mda olabilir ve aykÄ±rÄ± deÄŸerler azdÄ±r.")
    else:
        print("DeÄŸiÅŸim katsayÄ±sÄ± %35'in Ã¼zerinde. Veri aykÄ±rÄ± deÄŸerler bulunmaktadÄ±r.")
```

```python
def detect_outliers_iqr(data, column_name, Q1=0.25, Q3=0.75):
    # AykÄ±rÄ± deÄŸerleri tespit etmek iÃ§in IQR yÃ¶ntemini kullanma
    Q1 = df[column_name].quantile(Q1)
    Q3 = df[column_name].quantile(Q3)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)]
    
    # AykÄ±rÄ± deÄŸerleri kÃ¼Ã§Ã¼kten bÃ¼yÃ¼ÄŸe sÄ±ralama
    sorted_outliers = outliers.sort_values(by = column_name, ascending=False)
    
    print("AykÄ±rÄ± DeÄŸerler (KÃ¼Ã§Ã¼kten BÃ¼yÃ¼ÄŸe):")
    print(sorted_outliers[[column_name]].head(16))
    
    # AykÄ±rÄ± deÄŸerlerin sayÄ±sÄ±nÄ± bulma
    num_outliers = len(outliers)
    
    print("AykÄ±rÄ± DeÄŸerlerin SayÄ±sÄ±:")
    print(num_outliers)
    
    # AykÄ±rÄ± deÄŸerlerin benzersiz (unique) deÄŸerlerini alma
    unique_outliers = sorted_outliers[column_name].unique()
    
    print("Benzersiz AykÄ±rÄ± DeÄŸerler:")
    print(unique_outliers)
```

```python
def statistics_numerical_column(df, column_name):
    """
    Verilen sayÄ±sal sÃ¼tunun Ã¶zet istatistiklerini hesaplar ve sonucu normal biÃ§imde gÃ¶sterir.
    
    NOT: EÄŸer CV deÄŸeri yÃ¼zde 35'in altÄ±nda ise, veri setinin normal daÄŸÄ±lÄ±ma daha yakÄ±n olabileceÄŸi
    ve aykÄ±rÄ± deÄŸerlerin az olduÄŸu sÃ¶ylenebilir. 
    EÄŸer CV deÄŸeri yÃ¼zde 35'in Ã¼zerinde ise, veri setinde bÃ¼yÃ¼k bir deÄŸiÅŸkenlik olduÄŸu 
    ve aykÄ±rÄ± deÄŸerlerin bulunma olasÄ±lÄ±ÄŸÄ±nÄ±n yÃ¼ksek olduÄŸu anlamÄ±na gelir.
    
    Args:
    df : Veri seti
    column_name : Ä°ÅŸlenecek sÃ¼tunun adÄ±
    
    """
   
    # Verilen sÃ¼tunun Ã¶zet istatistiklerini gÃ¶sterme
    column_description = df[[column_name]].describe()

    print(f"'{column_name}' sÃ¼tununun Ã¶zet istatistikleri:")
    print(column_description)

    # mean ve std deÄŸerlerini Ã§ekmek iÃ§in doÄŸru indeksleme
    count = column_description.loc['count'][column_name]
    mean = column_description.loc['mean'][column_name]
    std = column_description.loc['std'][column_name]
    max_ = column_description.loc['max'][column_name]
    min_ = column_description.loc['min'][column_name]
    percentile_25 = column_description.loc['25%'][column_name]
    percentile_75 = column_description.loc['75%'][column_name]
    percentile_50 = column_description.loc['50%'][column_name]

    # DeÄŸiÅŸim katsayÄ±sÄ±nÄ± (CV) hesaplama
    cv = (std / mean) * 100  # YÃ¼zde olarak ifade edilir
    
    print(f"\nCount: {count:,.0f}")
    print(f"Mean: {mean:,.0f}")
    print(f"Standard Deviation (Std): {std:,.0f}")
    print(f"Minimum (Min): {min_:,.0f}")
    print(f"25th Percentile: {percentile_25:,.0f}")
    print(f"50th Percentile (Median): {percentile_50:,.0f}")
    print(f"75th Percentile: {percentile_75:,.0f}")
    print(f"Maximum (Max): {max_:,.0f}")
    print(f"\nDeÄŸiÅŸim KatsayÄ±sÄ± (CV): {cv:.2f}%")


    # CV deÄŸerini kontrol etme
    if cv < 35:
        print("DeÄŸiÅŸim katsayÄ±sÄ± %35'in altÄ±nda. Veri normal daÄŸÄ±lÄ±mda olabilir ve aykÄ±rÄ± deÄŸerler azdÄ±r.")
    else:
        print("DeÄŸiÅŸim katsayÄ±sÄ± %35'in Ã¼zerinde. Veri aykÄ±rÄ± deÄŸerler bulunmaktadÄ±r.")
```

```python
def plot_stacked_bar(df,column1,column2,rotation=0):
    print(f'\n{column1} & {column2} DaÄŸÄ±lÄ±mÄ±')
    palette = "deep" 
    sns.set_palette(palette)

    pd.crosstab(df[column1], df[column2]).plot(kind='bar', stacked=True)
    
    plt.xlabel(f'{column1}')
    plt.ylabel('KayÄ±t SayÄ±sÄ±')
    plt.title(f'{column1} & {column2} DaÄŸÄ±lÄ±mÄ±')
    plt.xticks(rotation=rotation)

    plt.show()
```

```python
def numeric_cleaning(df, col, groupby):    
    df_dropped = df[df[col].notna()].groupby(groupby)[col].apply(list)

    # GruplandÄ±rÄ±lmÄ±ÅŸ 'N_Age' listeleri Ã¼zerinde mode (mod) hesapla 
    # x'e min'i y'ye max'Ä± ata 
    x, y = df_dropped.apply(lambda x: stats.mode(x, keepdims=True)).apply([min, max])

    # Her bir mÃ¼ÅŸterinin 'N_Age' listesi iÃ§in minimum ve maksimum mod deÄŸerlerini al
    minn, maxx = x[0][0], y[0][0]
    
    # YanlÄ±ÅŸ deÄŸerlere Nan ata
    df[col] = df[col].apply(lambda x: np.nan if ((x<minn)|(x>maxx)|(x<0)) else x)
    
    # Local Mode ile Nan doldur
    mode_by_group = df.groupby(groupby)[col].transform(lambda x: x.mode()[0] if not x.mode().empty else np.NaN)
    df[col] = df[col].fillna(mode_by_group)
    
    print("\nTemizlendikten sonra Min, Max deÄŸerleri:", df[col].apply([min, max]), sep='\n', end='\n') 
    print("\nTemizlendikten sonra eÅŸsiz deÄŸerlerin sayÄ±sÄ±:",df[col].nunique())
    print("\nTemizlendikten sonra boÅŸ deÄŸer sayÄ±sÄ±:",df[col].isnull().sum())
```

     âš« "N_Type_of_Loan" sÃ¼tunundaki tÃ¼m deÄŸerler incelendi ve bu deÄŸerlerin iÃ§indeki benzersiz alt         deÄŸerler (loan <br>tÃ¼rlerini) bulundu.
```python
# Benzersiz loan tÃ¼rlerini saklamak iÃ§in boÅŸ bir set oluÅŸturuldu
unique_values = set()

""" 
    Ek Bilgi: 
    ----------------------------------------
    "append" metodu:
    KullanÄ±m AlanÄ±: Listeler iÃ§in kullanÄ±lÄ±r.
    Fonksiyon: Tek bir Ã¶ÄŸeyi listeye ekler.
    
    ---------------------------------------
    "extend" metodu:
    KullanÄ±m AlanÄ±: Listeler iÃ§in kullanÄ±lÄ±r.
    Fonksiyon: Bir listeye baÅŸka bir listenin Ã¶ÄŸelerini teker teker ekler.
    
    ---------------------------------------
    "update" metodu:
    KullanÄ±m AlanÄ±: Setler iÃ§in kullanÄ±lÄ±r.
    Fonksiyon: Bir iterable'dan (Ã¶rneÄŸin liste, tuple) Ã¶ÄŸeleri sete teker teker ekler.
    
    ----------------------------------------
    "lambda" ifadesi iÃ§inde " unique_values.update(x.split(",")) " ifadesi Ã§alÄ±ÅŸtÄ±rÄ±lÄ±yor.
     Bu ifade None dÃ¶ndÃ¼rÃ¼yor Ã§Ã¼nkÃ¼ set.update metodu bir deÄŸeri dÃ¶ndÃ¼rmez; sadece seti gÃ¼nceller.
     Bu nedenle, lambda ifadesinin dÃ¶ndÃ¼rdÃ¼ÄŸÃ¼ sonuÃ§ None olur.

"""
# dropna() metodu ile N_Type_of_Loan sÃ¼tunundaki NaN deÄŸerler atlanÄ±r.
# apply metodu ile her bir hÃ¼credeki deÄŸer, lambda fonksiyonuna geÃ§irilir.
# Lambda fonksiyonunda x.split(",") ifadesi ile hÃ¼credeki deÄŸer virgÃ¼lle ayrÄ±lÄ±r ve bir liste dÃ¶ner.
# strip() metodu ile listenin her bir Ã¶ÄŸesindeki saÄŸ ve sol boÅŸluklar temizlenir.
# unique_values.update() metodu ile bu temizlenmiÅŸ Ã¶ÄŸeler unique_values setine eklenir.

df["N_Type_of_Loan"].dropna().apply(lambda x: unique_values.update([item.strip() for item in x.split(",")]))
```

```python
# BoÅŸ deÄŸeri (empty string) 'unique_values' setinden Ã§Ä±karÄ±ldÄ±
unique_values.discard('')

# 'unique_values' setini bir listeye dÃ¶nÃ¼ÅŸtÃ¼rÃ¼ldÃ¼
unique_values_list = list(unique_values)
```

```python
""" 
    for value in unique_values_list:: unique_values_list iÃ§erisindeki her deÄŸer iÃ§in dÃ¶ngÃ¼ baÅŸlatÄ±lÄ±r.
    
    df[value] = df["N_Type_of_Loan"].apply(...): N_Type_of_Loan sÃ¼tununda her bir hÃ¼creyi kontrol 
    ederek, value deÄŸerini iÃ§eren hÃ¼creler iÃ§in yeni bir sÃ¼tun oluÅŸturur ve
    bu sÃ¼tuna 1, iÃ§ermeyenlere 0 atar.

    lambda x: 1 if pd.notna(x) and value in [item.strip() for item in x.split(",")] else 0: 
    Lambda fonksiyonu, hÃ¼credeki deÄŸerin NaN olup olmadÄ±ÄŸÄ±nÄ± ve value iÃ§eriÄŸini kontrol eder. 
    EÄŸer hÃ¼cre NaN deÄŸilse ve value iÃ§eriÄŸi bulunuyorsa 1, aksi takdirde 0 dÃ¶ner.
    HÃ¼credeki deÄŸeri virgÃ¼lle ayÄ±rarak kontrol eder ve her bir Ã¶ÄŸenin baÅŸÄ±ndaki ve 
    sonundaki boÅŸluklarÄ± temizler.

"""
for value in unique_values_list:
    df[value] = df["N_Type_of_Loan"].apply(lambda x: 1 if pd.notna(x) and value in [item.strip() for item in x.split(",")] else (0 if pd.notna(x) else np.nan))

```

# Ki Kare BaÄŸÄ±msÄ±zlÄ±k Testi
Ä°ki kategorik deÄŸiÅŸken arasÄ±ndaki baÄŸÄ±msÄ±zlÄ±ÄŸÄ± test etmek iÃ§in kullanÄ±lÄ±r.

```python
import scipy.stats as ss  # scipy.stats kÃ¼tÃ¼phanesini ss olarak iÃ§e aktar

# 1. AdÄ±m: Kategorik deÄŸiÅŸkenlerin seÃ§imi
cat_column = ["Month", "Category_Age", "N_Occupation", "N_Type_of_Loan", "Payment_of_Min_Amount",
              "N_Payment_Behaviour", "N_Credit_Mix", "Credit_Status"]

# 2. AdÄ±m: P-DeÄŸerlerini Ä°Ã§eren BoÅŸ Bir Matris OluÅŸturma
p_matrix = pd.DataFrame(index=cat_column, columns=cat_column)


for col1 in cat_column:
    for col2 in cat_column:
        if col1 == col2:
            p_matrix.loc[col1, col2] = np.nan  # Kendisiyle karÅŸÄ±laÅŸtÄ±rma yok
        else:
            contingency_table = pd.crosstab(dnew[col1], dnew[col2])
            # HÃ¼cre beklenen frekanslarÄ±nÄ±n kontrolÃ¼
            chi2, p, dof, ex = ss.chi2_contingency(contingency_table)
            if np.all(ex >= 5):
                p_matrix.loc[col1, col2] = p
            else:
                p_matrix.loc[col1, col2] = np.nan  # Beklenen frekanslar yeterli deÄŸilse NaN ata
  

# P-DeÄŸerlerini SayÄ±sal DeÄŸere DÃ¶nÃ¼ÅŸtÃ¼rme
p_matrix = p_matrix.astype(float)


# P-DeÄŸerlerini GÃ¶rselleÅŸtirme
plt.figure(figsize=(10, 8))
sns.heatmap(p_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Categorical Features Chi-Square Test p-Values')
plt.show()
```

```python
from scipy.stats import mannwhitneyu
# SayÄ±sal deÄŸiÅŸkenleri seÃ§me
numerical_columns = dnew.select_dtypes(include=[np.number]).columns

# P-deÄŸerlerini saklamak iÃ§in boÅŸ bir DataFrame oluÅŸturma
p_values = pd.DataFrame(index=numerical_columns, columns=['p_value'])

# Her sayÄ±sal deÄŸiÅŸken iÃ§in Mann-Whitney U Testi uygulama
for col in numerical_columns:
    group1 = dnew[dnew['Credit_Status'] == 'yes'][col]
    group2 = dnew[dnew['Credit_Status'] == 'no'][col]
    stat, p = mannwhitneyu(group1, group2)
    p_values.loc[col, 'p_value'] = p

# P-deÄŸerlerini sayÄ±sal hale getirme
p_values = p_values.astype(float)

# P-deÄŸerlerini Ä±sÄ± haritasÄ± olarak gÃ¶rselleÅŸtirme
plt.figure(figsize=(8, 6))
sns.heatmap(p_values, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Mann-Whitney U Testi P-DeÄŸerleri IsÄ± HaritasÄ±')
plt.show()

print(p_values)
```
